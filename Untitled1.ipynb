```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tocenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tocenizer.eos_token_id)
```

```python
text = 'What is the fastest car in the world?'
input_ids = tocenizer.encode(text, return_tensors='pt')
```

```python
display(input_ids)
```

```python
tocenizer.decode(input_ids[0][3])
```

```python
output = model.generate(input_ids, max_length=500, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

```python
tocenizer.decode(output[0], skip_special_tokens=True)
```
